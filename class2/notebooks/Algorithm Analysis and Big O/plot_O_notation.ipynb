```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
x = np.linspace(1,10**2,100)

fig = plt.figure(figsize=(12,6), dpi=200)
ax = fig.add_subplot(111)
ax.set_ylim(0, 10**4)
ax.plot(x, x, label='n')
ax.plot(x, x**2, label='$n^2$')
ax.plot(x, x**3, label='$n^3$')
ax.plot(x, x**4, label='$n^4$')
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles, labels)
```

```python
x = np.linspace(1,10**3,100)

fig = plt.figure(figsize=(12,6), dpi=200)
ax = fig.add_subplot(111)
ax.set_ylim(0, 10**3)
ax.plot(x, np.log2(x), label='$log_2 n$')
ax.plot(x, x*np.log2(x), label='$n log_2 n$')
ax.plot(x, x**2, label='$n^2$')
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles, labels)
```

```python
x = np.linspace(1,2*10**2,100)

fig = plt.figure(figsize=(12,6), dpi=200)
ax = fig.add_subplot(111)
ax.set_ylim(0, 2*10**6)
ax.plot(x, 100*x**2, label='$n^2$')
ax.plot(x, x**3, label='$n^3$')
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles, labels)
```
